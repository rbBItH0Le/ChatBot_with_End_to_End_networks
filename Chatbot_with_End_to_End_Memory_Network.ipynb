```python
vocab=set()
for story,question,answer in all_data:
    vocab=vocab.union(set(story))
    vocab=vocab.union(set(question))
```

```python
vocab.add('yes')
vocab.add('no')
```

```python
vocab_len=len(vocab)+1
```

```python
story_len=[len(data[0]) for data in all_data]
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer(filters=[])
```

```python
tokenizer.fit_on_texts(vocab)
```

```python
len(tokenizer.word_index)
```

```python
train_story_text=[]
train_question_text=[]
train_answer_text=[]
```

```python
for data,question,answer in train:
    train_story_text.append(data)
    train_question_text.append(question)
    train_answer_text.append(answer)
```

```python
train_question_text
```

```python
train_answer_text
```

```python
train_story_seq=tokenizer.texts_to_sequences(train_story_text)
```

```python
train_story_seq
```

```python
len(train_story_seq)
```

```python
def vectorize_stories(data,word_index,max_story_len,max_question_len):
    X=[]
    Xq=[]
    Y=[]
    
    for story,question,answer in data:
        x=[word_index[word.lower()] for word in story]
        q=[word_index[word.lower()] for word in question]
        y=np.zeros(len(word_index)+1)
        y[word_index[answer]]=1
        
        X.append(x)
        Xq.append(q)
        Y.append(y)
        
    return(pad_sequences(X,max_story_len),pad_sequences(Xq,max_question_len),np.array(Y))
```

```python
wordo=tokenizer.word_index
input_train,question_train,answer_train=vectorize_stories(train,wordo,max_story_len,max_question_len)
```

```python
input_train
```

```python
sum(answer_train)
```

```python
len(answer_train)
```

```python
input_test,question_test,answer_test=vectorize_stories(test,wordo,max_story_len,max_question_len)
```

```python
sum(answer_test)
```

```python
import tensorflow.keras as tf
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.models import Model
```

```python
from tensorflow.keras.layers import Embedding,Input,Activation,Dense,Permute,dot,add,concatenate,LSTM,Dropout
```

```python
input_sequence=Input((max_story_len,))
```

```python
question=Input((max_question_len,))
```

```python
vocab_size=len(vocab)+1
```

```python
vocab_size
```

```python
input_encoder_m=Sequential()
input_encoder_m.add(Embedding(input_dim=vocab_size,output_dim=64))
input_encoder_m.add(Dropout(0.3))
```

```python
input_encoder_c=Sequential()
input_encoder_c.add(Embedding(input_dim=vocab_size,output_dim=max_question_len))
input_encoder_c.add(Dropout(0.3))
```

```python
question_encoder=Sequential()
question_encoder.add(Embedding(input_dim=vocab_size,output_dim=64,input_length=max_question_len))
question_encoder.add(Dropout(0.3))
```

```python
question_encoder.summary()
```

```python
max_question_len
```

```python
input_encoder_m.summary()
```

```python
input_encoded_m=input_encoder_m(input_sequence)
input_encoded_c=input_encoder_c(input_sequence)
question_encoded=question_encoder(question)
```

```python
input_encoded_c
```

```python
input_encoded_m
```

```python
question_encoded
```

```python
match=dot([input_encoded_m,question_encoded],axes=(2,2))
```

```python
match=Activation('softmax')(match)
```

```python
response=add([match,input_encoded_c])
```

```python
response.shape
```

```python
response=Permute((2,1))(response)
```

```python
response.shape
```

```python
answer=concatenate([response,question_encoded])
```

```python
answer.shape
```

```python
answer=LSTM(32)(answer)
```

```python
answer=Dropout(0.5)(answer)
```

```python
answer=Dense(vocab_size)(answer)
```

```python
answer=Activation('softmax')(answer)
```

```python
model=Model([input_sequence,question],answer)
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history=model.fit([input_train,question_train],answer_train,batch_size=32,epochs=100,validation_data=([input_test,question_test],answer_test))
```

```python
pred_results=model.predict([input_test,question_test])
```

```python
val_max=np.argmax(pred_results[0])
```

```python
for key,val in tokenizer.word_index.items():
    if val==val_max:
        print(key)
```
